```python
import matplotlib. pyplot as plt
import numpy as np
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
xpoints=np.array([5,22,13.03,10,10,20,30,40])
ypoints=np.array([5,39,36.46,5,20,30,40,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib. pyplot as plt
import numpy as np
a1=[5,22,13.03,10,10,20,30,40]
a2=[5,39,36.46,5,20,30,40,100]
plt.bar(a1,a2)
plt.title("a1 vs a2")
plt.xlabel("a1")
plt.ylabel("a2")
plt.show()
```

```python
import matplotlib. pyplot as plt
import numpy as np
plt.title("pie")
x=np.array([5,22,13,10,10,20,30,40])
co=["r","b","hotpink","y","g","v","black","white"]
la=["count","mean","std","min","25%","50%","75%","max"]
plt.pie(x,colors=co,labels=la)
plt.show()
```
