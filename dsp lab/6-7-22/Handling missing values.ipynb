```python
import pandas as pd
import numpy as np
student=pd.DataFrame({"NAME":["lakshmi","USHA","Venkateswari","Naveena","vara lakshmi","navya","vanaja","bhuvana","usha sri"],"ID":[np.nan,567,133,np.nan,123,567,768,np.nan,987],"MARKS":[89,92,86,89,86,np.nan,96,90,85],"GROUP":["cse","ece","ece","eee",np.nan,"mech","civil",np.nan,"civil"]})
```

```python
student
```

```python
student1=student.copy()
column="ID"
student1["ID"]=(student1["ID"].min())/(student1["ID"].max()-student1["ID"].min())
display(student1)
```

```python
s3=student.copy()
column="MARKS"
s3["MARKS"]=(s3["MARKS"]/s3["MARKS"].abs().max())
display(s3)
```

```python
s4=student.copy()
column="ID"
s4["ID"]=(s4["ID"]-s4["ID"].mean())/s4["ID"].std()
display(s4)
```

```python
student["bins"]=pd.cut(x=student["ID"],bins=[567,133,123,567,768,987])
print(student)
```

```python
#removing the columns  between two indexes
student.drop(student.iloc[:,2:])
student
```
